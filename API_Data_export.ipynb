```python
# Import Dependencies
import requests
import json
import pandas as pd

# Import the API key
from config import eBird_api_key
from config import Google_maps_api_key

#Import pprint
from pprint import pprint
```

```python
url = f"https://api.ebird.org/v2/data/obs/AU/recent/notable?back=30&detail=full"
headers = {"X-eBirdApiToken": eBird_api_key}    

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    ebird_data = response.json()
except requests.exceptions.RequestException as e:
    print(f"Failed to fetch data: {e}")
```

```python
pprint(ebird_data)
```

```python
obs_no = (len(ebird_data))
print(f"Number of observations in dataset: {obs_no}")
```

```python
# Create an empty list to store merged data
merged_data = []

# Iterate through each observation in the eBird API response
for observation in ebird_data:
    lat = observation['lat']
    lng = observation['lng']

    # Construct the URL for the reverse geocoding request
    url = f'https://maps.googleapis.com/maps/api/geocode/json?latlng={lat},{lng}&key={Google_maps_api_key}'

    # Make a GET request to the Google Maps Geocoding API
    geocoding_response = requests.get(url).json()

    # Extract the formatted address from the response
    if 'results' in geocoding_response and len(geocoding_response['results']) > 0:
        formatted_address = geocoding_response['results'][0]['formatted_address']

        # Create a new dictionary combining eBird observation data and formatted address
        merged_observation = observation.copy()  # Copy all original observation data
        merged_observation['formatted_address'] = formatted_address  # Add the formatted address

        # Append the merged observation to the list
        merged_data.append(merged_observation)

# Pretty print the merged data for inspection
pprint(merged_data)
```

```python
merged_data_new = pd.DataFrame(merged_data)

# Function to clean formatted_address column
def clean_formatted_address(address):
    if '+' in address:
        return address.split('+')[-1].split(' ', 1)[-1]
    else:
        return address

# Apply the cleaning function to the formatted_address column
merged_data_new['formatted_address'] = merged_data_new['formatted_address'].apply(clean_formatted_address)
```

```python
# Check for data containing "-"
merged_data_new['formatted_address'] = merged_data_new['formatted_address'].apply(lambda x: x.split("-")[0] if "-" in x else x)

# Remove all values after the first comma
merged_data_new['formatted_address'] = merged_data_new['formatted_address'].apply(lambda x: x.split(",")[0])

# Add the values of "subnational1Code" separated by commas
merged_data_new['formatted_address'] = merged_data_new['formatted_address'] + ', ' + merged_data_new['subnational1Code']
```

```python
# Function to replace values with 2 characters in formatted_address with locName
def replace_short_values(row):
    if len(row['formatted_address']) == 2:
        return row['locName']
    else:
        return row['formatted_address']

# Apply the function to the formatted_address column
merged_data_new['formatted_address'] = merged_data_new.apply(replace_short_values, axis=1)
```

```python
# List of substrings to remove from the beginning of the address
substrings_to_remove = ["VIC", "NSW", "ACT", "WA", "NT", "SA", "TAS"]

# Function to remove substrings from the beginning of a string before the first comma
def remove_substrings(text):
    parts = text.split(",", 1)
    address = parts[0].strip()
    for substring in substrings_to_remove:
        if address.startswith(substring):
            address = address[len(substring):].lstrip()
    return address

# Function to remove characters after the first comma that start with "AU-"
def remove_characters_after_comma(text):
    parts = text.split(",", 1)
    if len(parts) > 1 and parts[1].strip().startswith("AU-"):
        return parts[0]
    return text

# Apply the functions to clean the formatted_address column
merged_data_new['formatted_address'] = merged_data_new['formatted_address'].apply(lambda x: remove_characters_after_comma(remove_substrings(x)))
```

```python
# Function to add missing state codes to the formatted_address
def add_missing_state_codes(row):
    state_codes = ["VIC", "NSW", "ACT", "WA", "NT", "SA", "TAS","BNE","QLD"]
    for state_code in state_codes:
        if state_code in row['subnational1Code'] and state_code not in row['formatted_address']:
            row['formatted_address'] += f", {state_code}"
    return row['formatted_address']

# Apply the function to add missing state codes to the formatted_address column
merged_data_new['formatted_address'] = merged_data_new.apply(add_missing_state_codes, axis=1)
```

```python
# Function to update locName with values from formatted_address
def update_locName_with_formatted_address(row):
    row['locName'] = row['formatted_address']
    return row

# Apply the function to update locName with formatted_address values
merged_data_new = merged_data_new.apply(update_locName_with_formatted_address, axis=1)

# Display the updated DataFrame
print(merged_data_new)
```

```python
# Delete the 'formatted_address' category from the DataFrame
if 'formatted_address' in merged_data_new.columns:
    merged_data_new = merged_data_new.drop('formatted_address', axis=1)

# Display the updated DataFrame
print(merged_data_new)
```

```python
# Convert the DataFrame to a dictionary
merged_data_dict = merged_data_new.to_dict(orient='records')

# Serialize the cleaned data (now as a dictionary) to JSON format with an indentation of 4 spaces
json_data = json.dumps(merged_data_dict, indent=4)

# Print the serialized JSON data
```
